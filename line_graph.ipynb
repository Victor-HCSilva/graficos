# Line Graph

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Data
x = [- 1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Configs 
color='blue'
linestyle='-'
marker='o'
title = "Line graph"
x_title = "Eixo X"
y_title = "Eixo Y"

# graph
plt.plot(x, y, color=color, linestyle=linestyle, marker=marker)

# Adicionar títulos e rótulos
plt.title(title)
plt.xlabel(x_title)
plt.ylabel(y_title)
# plt.xlim(left=0)
# plt.ylim(bottom=0)
# Exibir o gráfico
plt.show()
```

# Box Plot 

```python
# Dados de exemplo
dados = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Criar o gráfico
plt.boxplot(dados)
plt.title('Exemplo de Boxplot Simples')
plt.ylabel('Valores')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```

# Histogram

```python
# 1. Dados
dados = np.random.randn(1000) # Exemplo: 1000 valores aleatórios

# 2. Criar Histograma
plt.hist(dados, bins=30, color='blue', edgecolor='black', alpha=0.7)

# 3. Personalizar
plt.title('Histograma Simples')
plt.xlabel('Valores')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.5)
# plt.xlim(left=0)
# plt.ylim(bottom=0)
# 4. Exibir
plt.show()
```

# Curva implícita


```python

x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 4, 400)

X, Y = np.meshgrid(x, y)

Z = (X**3 - 1)**2 - (Y - 2)**2 - 2

plt.contour(X, Y, Z, levels=[0])
plt.title("Curva implícita")
plt.grid()
# plt.xlim(left=0)
# plt.ylim(bottom=0)
plt.show()
```

# Circunfêrencia

```python
a, b = 0, 0   # centro
r = 2         # raio

x = np.linspace(-2, 2, 1000)
y = np.sqrt(r**2 - (x - a)**2)

plt.plot(x, y)     # parte de cima
plt.plot(x, - 1.36 * y)    # parte de baixo

plt.gca().set_aspect('equal')  # mantém o círculo redondo
plt.grid()
plt.title("Círculo (forma explícita)")
# plt.xlim(left=0)
# plt.ylim(bottom=0)
plt.show()
```
